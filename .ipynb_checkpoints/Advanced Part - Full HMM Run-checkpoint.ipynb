```python
import pandas as pd
import sklearn
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from sklearn.metrics import recall_score, precision_score, f1_score
from sklearn.preprocessing import MinMaxScaler
from hmmlearn import hmm
import copy
import time
import math
```

```python
def Generate_test(path):
    df_train = pd.read_csv("Weather_Data/"+path.format('train'))
    number_bars = 20
    
    df_test = pd.read_csv("Weather_Data/"+path.format('test'))
    columns_to_hmm = [' _dewptm_bars', ' _fog', ' _hail', ' _hum_bars', ' _pressurem_bars',' _rain', ' _snow', ' _thunder', ' _tornado', ' _vism_bars', ' _wspdm_bars']
    other_columns = [' _conds','datetime','year','month','week','day','night','morning','noon','evening', 'season_summer', 'season_winter', 'season_spring', 'season_monsoon', 'season_autumn', 'month_cos', 'month_sin', 'week_cos', 'week_sin']
    results = pd.DataFrame(df_test['Temp'].tolist(), columns = ['Temp'])
    
    for c in other_columns:
        c_list = df_test[c].tolist()
        c_list = np.array(c_list)
        results[c] = c_list
        columns = df_train.columns
    for c in columns_to_hmm:
        print(c)
        values_list = df_train[c].tolist()
        #print(values_list)
        num_of_uinque_states = len(set(values_list))
        print('unique',num_of_uinque_states)
        last_value = values_list[-1]
        #print(values_list)
        values_list = [[v] for v in values_list]
        #print(values_list)
        model = hmm.GaussianHMM(n_components=num_of_uinque_states, covariance_type="full")
        model.fit(values_list)
        #model.startprob_ = find_start_state(model,last_value,num_of_uinque_states)
        print(df_test[c].size)
        X, Z = model.sample(1+df_test[c].size)
        X = [item for sublist in X for item in sublist]
        X = np.round(X)
        print(len(X))
        results[c] = X[1:]
        
    results.to_csv('HMM_'+path.format('test'))

    
```

```python
for freq_range in [[str(i) for i in range(1, 13)],  # month
                       ['autumn', 'winter', 'spring', 'monsoon', 'summer']]:  # season
                                  
    for freq_part in freq_range:
        for day_part in ['night', 'morning', 'noon', 'evening']:
            print() 
            Generate_test('_'.join(['df',freq_part, day_part,'{}.csv'])) 
```
